Example of match: Search the string to see if it starts with "Python"

```python
import re
```

```python
msg= "Python is an object-oriented language"
```

```python
x =re.match("Python",msg)
if(x):
  print("Pattern is matched")
else:
  print("No match")
```

Example of findall(): function returns a list containing all matches
Example: Print a list of all matches:

```python
txt="The rain in Spain"
x=re.findall('ai',txt)
print(x)
```

The Search() Function
Example: Make a search that return no match:

```python
txt="The rain in Spain"
x=re.search("Portugal",txt)
print(x)
```

The split() Function: The split function returns a list where the string has been split at each match:
Example: Split at each white-space character:

```python
txt="The rain in Spain"
x=re.split("\s",txt)
print(x)
```

Example2: Split the string only at the first occurance

```python
txt="The rain in Spain"
x=re.split("\s",txt,1)
print(x)
```

The search() function searches the string for a

match and returns a match object if there is a match.
If there is more than one match, only the first occurance of the match will be returned.

```python
# Example : Search for the first position of white-space character in the string
txt="The rain in Spain"
x=re.search("\s",txt)
print("The first white-space character is located in position:",x.start())
```

The sub() function replaces the matches with the text of your choice:
Example: Replace every white-space character with the number 9:

```python
txt="The rain in Spain"
x=re.sub("\s","9",txt)
print(x)
```

Note: You can control the number of replacements by specifying the count parameter.

```python
txt="The rain in Spain"
# Find all lower case characters alphabetically between "a" and "m":
x=re.findall("[a-m]",txt)
print(x)
```

```python
txt="That will be 59 dollars"
# Find all digits:
x=re.findall("\d",txt)
print(x)
```

```python
txt="hello planet"
# Check if the string starts with 'hello':
x=re.findall("^hello",txt)
if x:
  print("Yes, the string starts with 'hello'")
else:
  print("No match")
```

```python
txt="hello planet"
# Check if the string ends with 'planet':
x=re.findall("planet$",txt)
if x:
  print("Yes, the string end with 'planet'")
else:
  print("No match")
```

```python
txt="hello planet"
# Search for a sequence that starts with "he", followed by o or more (any) characters, and "o":
x=re.findall("he.*o",txt)          * means any character
print(x)
```

```python
txt="hello planet"
# Search for a sequence that starts with "he", followed by 1 or more (any) characters, and "o":
x=re.findall("he.+o",txt)
print(x)
```

```python
txt="hello planet"
# Search for a sequence that starts with "he", followed by 0 or 1 (any) characters, and an "o":
x=re.findall("he.?o",txt)
print(x)
#Note: This time we got no match, because there were not zero, not one, but two characters between "he" and the "o"
```

```python
txt="hello planet"
# Search for a sequence that starts with "he", followed by exactly 2 (any) characters, and an "o":
x=re.findall("he.{2}o",txt)
print(x)
```

```python
txt="hello planet"
# Search for a sequence that starts with "he", followed by two (any) characters, and an "o":
x=re.findall("he..o",txt)
print(x)
```

```python
txt="The rain in Spain falls mainly in the plain!"
# Check if the string contains either "falls" or "stays"':
x=re.findall("falls|stays",txt)
print(x)
if x:
  print("Yes, there is at least one match|")
else:
  print("No match")
```

Search the string to see if it starts with "The" and ends with "Spain":

```python
txt="The rain in Spain"
# Check if the string starts with "The" and ends with "Spain":
x=re.search("^The.*Spain$",txt)
if x:
  print("Yes, we have a match match|")
else:
  print("No match")
```
